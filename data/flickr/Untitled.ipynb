```python
import networkx as nx
```

```python
import json
from networkx.readwrite import json_graph
```

```python
jsondata = json.load(open('./flickr-G-formatted.json'))
G = json_graph.node_link_graph(jsondata)
```

```python
contents = []
for node in G.nodes():
    content = G.nodes[node]['content']
    if len(content)==1:
        contents.append(content[0])
    else:
        print(content)
```
